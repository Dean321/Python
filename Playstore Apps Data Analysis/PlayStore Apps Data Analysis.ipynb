```python
data_clean = data[data["Price"]<250]
data_clean.sort_values("Price", ascending=False)
```

```python
data_clean["Revenue_Estimate"] = data_clean.Installs.mul(data_clean.Price)
data_clean.sort_values("Revenue_Estimate", ascending=False)[:10]
```

```python
cat_install = data_clean.groupby("Category").agg({"Installs":pd.Series.sum})
cat_install.sort_values("Installs", ascending=True, inplace=True)
```

```python
h_bar = px.bar(x = cat_install.Installs,
              y = cat_install.index,
              orientation="h",
              title="Category Popularity")
h_bar.update_layout(xaxis_title="#Downloads", yaxis_title="Category")
h_bar.show()
```

```python
cat_num = data_clean.groupby("Category").agg({"App":pd.Series.count})
cat_merge_data = pd.merge(cat_num, cat_install, on="Category", how="inner")
print(cat_merge_data.shape)
cat_merge_data.sort_values("Installs",ascending=False)
```

```python
scatter = px.scatter(cat_merge_data,
                    x="App",
                    y="Installs",
                    title="Category Concentration",
                    size="App",
                    hover_name=cat_merge_data.index,
                    color="Installs")
scatter.update_layout(xaxis_title="#Apps",
                     yaxis_title="Installs",
                     yaxis=dict(type="log"))
scatter.show()
```

```python
box1 = px.box(
    data_clean,
    y="Installs",
    x="Type",
    color="Type",
    notched=True,
    points="all",
    title="How many downloads are paid apps giving up?"
)
box1.update_layout(yaxis=dict(type="log"))
box1.show()
```

From the hover text in the chart, we see that the median number of downloads for free apps is `100,000`, while the median number of downloads for paid apps is around `1,000`! This is massively lower.

```python
paid_apps = data_clean[data_clean["Type"] == "Paid"]
box2 = px.box(paid_apps,
             x="Category",
              y="Revenue_Estimate",
              title="How uch can paid apps earn?"
)
box2.update_layout(xaxis_title="Category",
            yaxis_title="Paid App ballpark revenue",
                   xaxis={"categoryorder":"min ascending"},
                   yaxis=dict(type="log")
)
box2.show()
```

If an Android app costs `$30,000` to develop, then the average app in very few categories would cover that development cost. The median paid photography app earned about `$20,000`. Many more app’s revenues were even lower - meaning they would need other sources of revenue like advertising or in-app purchases to make up for their development costs. However, certain app categories seem to contain a large number of outliers that have much higher (estimated) revenue - for example in `Medical`, `Personalisation`, `Tools`, `Game`, and `Family`.

```python

box3 = px.box(paid_apps,
             x="Category",
              y="Price",
              title="Price per Category"
)
box3.update_layout(xaxis_title="Category",
            yaxis_title="Paid App Price",
                   xaxis={"categoryorder":"max descending"},
                   yaxis=dict(type="log")
)
box3.show()
```

```python
paid_apps.Price.median()
```

However, some categories have higher median prices than others. This time we see that Medical apps have the most expensive apps as well as a median price of `$5.49`. In contrast, Personalisation apps are quite cheap on average at `$1.49`. Other categories which higher median prices are Business `($4.99)` and Medical `($4.99)`. It seems like customers who shop in these categories are not so concerned about paying a bit extra for their apps.
